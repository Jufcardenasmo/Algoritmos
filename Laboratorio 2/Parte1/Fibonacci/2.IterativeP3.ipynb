# Secuencia de Fibonacci en Python.

"The function time.time() returns the current system time in ticks since 12:00am, January 1, 1970(epoch)."
https://www.tutorialspoint.com/python3/python_date_time.htm

Se calcula el tiempo al iniciar y finalizar de cada ciclo. Luego se restan estos valores y se obtiene la duraci√≥n del ciclo.

```python
from matplotlib import pyplot
import time 

def fibonacci(n):
  ant = 0
  pre = 1
  sig = 1
  
  if int(n) == 0:
    return 0
  elif int(n) == 1:
    return 1
  
  inicio = time.time()
  for i in range(n):
    sig = ant + pre
    ant, pre = pre, sig
  final = time.time()
  duracion = final-inicio
  return ('%.4f' % duracion)

    
x = range(0, 10**5, 5000)
pyplot.plot(x, [fibonacci(i) for i in x], marker='o')

pyplot.axhline(0, color="black")
pyplot.axvline(0, color="black")

pyplot.xticks([i for i in range(0, 10**5 + 10**4, 2*10**4)])

pyplot.title(" Tiempo de ejecuci√≥n en funci√≥n de n")
pyplot.xlabel("n")  
pyplot.ylabel("Tiempo (s)") 

pyplot.savefig("fiboPlot4.png")
pyplot.show()


```
